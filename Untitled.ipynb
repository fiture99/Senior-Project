```python
#hide
from fastbook import *
from fastai.vision.widgets import *
```

```python
fish_types = 'Albacore tuna','Bigeye tuna','Bonga', 'Mahi Mahi', 'Opah','Other', ' Sharks', 'Yellowfin tuna'
path = Path('data/fish/train')
```

```python
if not path.exists():
    path.mkdir()
    for o in bear_types:
        dest = (path/o)
        dest.mkdir(exist_ok=True)
        results = search_images_bing(key, f'{o} bear')
        download_images(dest, urls=results.attrgot('content_url'))
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```

```python
class DataLoaders(GetAttr):
    def __init__(self, *loaders): self.loaders = loaders
    def __getitem__(self, i): return self.loaders[i]
    train,valid = add_props(lambda i,self: self[i])
```

```python
fish = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = fish.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
fish = fish.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls = fish.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
fish = fish.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode='zeros'))
dls = fish.dataloaders(path)
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
fish = fish.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = fish.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1, unique=True)
```

```python
fish = fish.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = fish.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```python
fish = fish.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = fish.dataloaders(path)
```

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
learn.fine_tune(5,5e-4)
```

```python
interp.plot_top_losses(5, nrows=1)
```
